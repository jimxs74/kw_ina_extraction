```python
import pandas as pd
import os
import sys

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
sys.path.append(PROJECT_ROOT)

from pke_ia.f_evaluation import evaluate_prediction
```

```python
# ambil data
path_data = os.path.join(PROJECT_ROOT, "data", "dataset_ekstraksi_r30.xlsx")
df_eval = pd.read_excel(path_data, sheet_name='Sheet1')

targets_list = df_eval[["k1", "k2", "k3","k4", "k5", "k6","k7"]].values.tolist()
predict_list = df_eval[['key_1','key_2','key_3']].values.tolist()

pke_tr_eval = evaluate_prediction(predict_list, targets_list)
```

```python
pke_tr_eval
```

```python
# Calculate Score, using flexible score : exact maatch =1, partial match = 1, no match = 0

pke_tr_recall = pke_tr_eval['flex_recall'].mean()
pke_tr_prec = pke_tr_eval['flex_prec'].mean()
pke_tr_f1 = 2 * (pke_tr_prec * pke_tr_recall) / (pke_tr_prec + pke_tr_recall)
```

```python
# Create a DataFrame with the scores
summary = pd.DataFrame({
    'pke_textrank': [pke_tr_recall, pke_tr_prec, pke_tr_f1],
}, index=['recall', 'precision', 'F1'])
summary =summary.round(3)
summary
```
